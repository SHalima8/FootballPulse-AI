```python
df = pd.read_csv('../data/raw/WorldCupMatches.csv')
print(f"Dataset loaded ✅")
print(f"Shape: {df.shape}")
```

```python
# Drop all rows where Home Team Name is empty
df = df.dropna(subset=['Home Team Name', 'Away Team Name'])

# Fix data types
df['Year'] = df['Year'].astype(int)
df['Home Team Goals'] = df['Home Team Goals'].astype(int)
df['Away Team Goals'] = df['Away Team Goals'].astype(int)

# Create Winner column
def get_winner(row):
    if row['Home Team Goals'] > row['Away Team Goals']:
        return row['Home Team Name']
    elif row['Away Team Goals'] > row['Home Team Goals']:
        return row['Away Team Name']
    else:
        return 'Draw'

df['Winner'] = df.apply(get_winner, axis=1)

# Verify
print(f"Clean dataset shape: {df.shape}")
print(f"Missing values remaining:\n{df.isnull().sum()}")
print(f"\nSample winners:\n{df['Winner'].value_counts().head(10)}")
```

```python
df.to_csv('../data/processed/clean_matches.csv', index=False)
print("Clean data saved to data/processed/ ✅")
```

```python
# Load new data
df_new = pd.read_csv('../data/raw/Match_2018_2022.csv')

# Load existing clean data
df_old = pd.read_csv('../data/processed/clean_matches.csv')

# Combine
df_combined = pd.concat([df_old, df_new], ignore_index=True)
df_combined = df_combined.sort_values('Year').reset_index(drop=True)

# Re-apply Winner column since new data derives it from goals
def get_winner(row):
    if row['Home Team Goals'] > row['Away Team Goals']:
        return row['Home Team Name']
    elif row['Away Team Goals'] > row['Home Team Goals']:
        return row['Away Team Name']
    else:
        return 'Draw'

df_combined['Winner'] = df_combined.apply(get_winner, axis=1)

print(f"Old data: {len(df_old)} matches")
print(f"New data: {len(df_new)} matches")
print(f"Combined: {len(df_combined)} matches")
print(f"Years: {df_combined['Year'].min()} → {df_combined['Year'].max()}")

# Save
df_combined.to_csv('../data/processed/clean_matches.csv', index=False)
print("Saved combined dataset ✅")
```
